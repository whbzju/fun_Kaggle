```python
def toCategory(cols, df_data):
    for name in cols:
        le = preprocessing.LabelEncoder()
        le.fit(df_data[name])
        tt = le.transform(df_data[name])
        df_data[name+'_cat'] = tt
```

```python
def transformNan(df):
    df.site_id[df.site_id == '85f751fd'] = np.nan
    df.site_domain[df.site_domain == 'c4e18dd6'] = np.nan
    df.site_category[df.site_category == '50e219e0'] = np.nan
    df.app_id[df.app_id == 'ecad2386'] = np.nan
    df.app_domain[df.app_domain == '7801e8d9'] = np.nan
    df.app_category[df.app_category == '07d7df22'] = np.nan
```

# 独立模型程序，不依赖notebook

```python
df_train_app = pd.read_csv('train_app_clean.csv')
df_train_app.head()
```

```python
df_train_pc = pd.read_csv('train_pc_clean.csv')
df_train_pc.head()
```

```python
def trainModel(df, key):
    clf = RandomForestClassifier(n_estimators=100, n_jobs=8, )
    if 'pc' in key:
        print "pc"
        clf.fit(df[['C1_cat', 'banner_pos_cat', 'site_id_cat',
       'site_domain_cat', 'site_category_cat', 'device_model_cat',
       'device_type_cat', 'device_conn_type_cat', 'C14_cat', 'C15_cat',
       'C16_cat', 'C17_cat', 'C18_cat', 'C19_cat', 'C20_cat', 'C21_cat']].values, df['click'].values)
    elif 'app' in key:
        print "app"
        clf.fit(df[['C1_cat', 'banner_pos_cat', 'app_id_cat',
       'app_domain_cat', 'app_category_cat', 'device_model_cat',
       'device_type_cat', 'device_conn_type_cat', 'C14_cat', 'C15_cat',
       'C16_cat', 'C17_cat', 'C18_cat', 'C19_cat', 'C20_cat', 'C21_cat']].values, df['click'].values)
    else:
        print "error data"
```

```python
clf = trainModel(df_train_pc, 'pc')

from sklearn.externals import joblib
joblib.dump(clf, 'model/rf100pc.pkl') 
```
