```python
df["Gender"] = df["Sex"].map({"female":0, "male":1})
df.head()
```

```python
df = df.dropna(axis=0)
df_test = df.loc[:, ("Pclass", "Age", "SibSp", "Parch", "Fare", "Gender", "Embarked_d")]
```

```python
df.isnull()
```

```python
df_test.isnull()
```

```python
df_label = df["Survived"]
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10)
clf.fit(df_test.values, df_label.values)
```

```python
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(clf, df_test.values, df_label.values) 
print scores
```
